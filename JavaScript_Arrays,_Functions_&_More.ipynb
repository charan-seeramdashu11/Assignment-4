2.Write a function checkEligibility that takes an age as a parameter
and returns:
● "Eligible" if age is 18 or above
● "Not Eligible" if age is below 18
Use an if-else statement.

````python
def checkEligibility(age):
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"
````

````python
# Example 1
age1 = 20
print(f"Age {age1}: {checkEligibility(age1)}")

# Example 2
age2 = 16
print(f"Age {age2}: {checkEligibility(age2)}")

# Example 3
age3 = 18
print(f"Age {age3}: {checkEligibility(age3)}")
````

3. Write a function getDayName that takes a number (1-7) and returns the
corresponding day of the week. Use a switch statement.

````python
def getDayName(day_number):
    # Using a dictionary to mimic a switch statement
    day_names = {
        1: "Monday",
        2: "Tuesday",
        3: "Wednesday",
        4: "Thursday",
        5: "Friday",
        6: "Saturday",
        7: "Sunday"
    }
    return day_names.get(day_number, "Invalid day number")
````

### Examples of getDayName function usage:

````python
# Example 1
print(f"Day 3 is: {getDayName(3)}")

# Example 2
print(f"Day 7 is: {getDayName(7)}")

# Example 3 (Invalid input)
print(f"Day 0 is: {getDayName(0)}")

# Example 4 (Invalid input)
print(f"Day 9 is: {getDayName(9)}")
````

5.Given an array numbers = [1, 2, 3, 4, 5], perform the following:
1. Use a for loop to print each number.
2. Use a while loop to print each number.
3. Use a for-in loop to print each index.
4. Use a for-of loop to print each number.

````python
numbers = [1, 2, 3, 4, 5]
print(f"Original array: {numbers}")
````

### 1. Use a `for` loop to print each number

````python
print("Printing numbers using a for loop:")
for number in numbers:
    print(number)
````

### 2. Use a `while` loop to print each number

````python
print("Printing numbers using a while loop:")
i = 0
while i < len(numbers):
    print(numbers[i])
    i += 1
````

### 3. Use a `for-in` loop to print each index

````python
print("Printing indices using a for-in (range) loop:")
for index in range(len(numbers)):
    print(index)
````

### 4. Use a `for-of` loop to print each number

````python
print("Printing numbers using a for-of (standard for) loop:")

for number in numbers:
    print(number)
````

6. Explain what functions are in JavaScript and provide an example of a
function that adds two numbers.

### Example of a JavaScript Function to Add Two Numbers

````python
print("```javascript")
print("// Function declaration")
print("function addNumbers(a, b) {")
print("  return a + b;")
print("}")
print("\n// Function expression (anonymous function)")
print("const subtractNumbers = function(a, b) {")
print("  return a - b;")
print("}")
print("\n// Arrow function (modern syntax)")
print("const multiplyNumbers = (a, b) => { // Or (a, b) => a * b for single expression")
print("  return a * b;")
print("}")
print("\n// Calling the function and storing the result")
print("let result = addNumbers(5, 3);")
print("console.log('5 + 3 = ' + result); // Output: 5 + 3 = 8")
print("\nconsole.log('10 - 4 = ' + subtractNumbers(10, 4)); // Output: 10 - 4 = 6")
print("console.log('6 * 7 = ' + multiplyNumbers(6, 7)); // Output: 6 * 7 = 42")
print("```")
````

9.Given the array:
const fruits = ["apple", "banana", "cherry"];
1. What is the length of the array?
2. Add "date" to the end of the array using the length property.

````python
# Define the Python list equivalent to the JavaScript array
fruits = ["apple", "banana", "cherry"]
print(f"Initial fruits array: {fruits}")
````

````python
array_length = len(fruits)
print(f"The length of the array is: {array_length}")
````

### 2. Add "date" to the end of the array using the length property.

````python
# In Python, the idiomatic way to add an element to the end is using .append().
# While JavaScript allows direct assignment to `array[array.length]`,
# Python lists use `len()` to get the length, and direct assignment to `list[len(list)]`
# would result in an IndexError if the index is out of bounds. `append()` is the correct method.
fruits.append("date")

print(f"Fruits array after adding 'date': {fruits}")
````

10.
 Given the string text = " Hello World! ", perform the following:
1. Trim the whitespace.
2. Convert to uppercase.
3. Check if it includes "World".

````python
text = " Hello World! "
print(f"Original string: '{text}'")
````

### 1. Trim the whitespace.

````python
trimmed_text = text.strip()
print(f"Trimmed string: '{trimmed_text}'")
````

### 2. Convert to uppercase.

````python
uppercase_text = trimmed_text.upper()
print(f"Uppercase string: '{uppercase_text}'")
````

### 3. Check if it includes "World".

````python
# Checking for 'World' in the original trimmed string (case-sensitive)
includes_world = "World" in trimmed_text
print(f"Does the trimmed string include 'World'? {includes_world}")

# Checking for 'WORLD' in the uppercase string
includes_world_upper = "WORLD" in uppercase_text
print(f"Does the uppercase string include 'WORLD'? {includes_world_upper}")
````

11.
Given the array numbers = [1, 2, 3, 4, 5], perform the following:
1. Add 6 to the end.
2. Remove the first element.
3. Find the index of 4.
4. Check if all numbers are greater than 0.

````python
numbers = [1, 2, 3, 4, 5]
print(f"Initial numbers array: {numbers}")
````

### 1. Add 6 to the end.

````python
numbers.append(6)
print(f"Numbers array after adding 6: {numbers}")
````

### 2. Remove the first element.

````python
if numbers:
    removed_element = numbers.pop(0) # Removes and returns the element at index 0
    print(f"Numbers array after removing the first element ({removed_element}): {numbers}")
else:
    print("Array is empty, cannot remove first element.")
````

### 3. Find the index of 4.

````python
try:
    index_of_4 = numbers.index(4)
    print(f"The index of 4 is: {index_of_4}")
except ValueError:
    print("4 is not in the array.")
````

### 4. Check if all numbers are greater than 0.

````python
all_greater_than_0 = all(num > 0 for num in numbers)
print(f"Are all numbers greater than 0? {all_greater_than_0}")
````
